```python
import math
def discriminant_root(numbers):
  numbers_list = numbers.split(' ')
  l = []
  for i in numbers_list:
    l.append(int(i))
  a = l[0]
  b = l[1]
  c = l[2]
  d = math.pow(b, 2) - 4 * a * c
  if d < 0:
    return f"Дискриминант меньше нуля"
  else:
    d_sqrt = math.sqrt(d)
    return d_sqrt



```
