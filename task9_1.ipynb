```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('books.csv')
col_year = df.columns[0]
col_month = df.columns[1]
col_books = df.columns[2]
col_pages = df.columns[3]
```

```python
def create_line_chart_2023():

    data_2023 = df[df[col_year] == 2023]

    plt.plot(data_2023[col_month],
             data_2023[col_books],
             marker='o',
             color='red',
             label='Прочитано книг')

    plt.title('Прочитанные книги 2023')

    plt.xlabel('Месяцы')
    plt.ylabel('Количество книг')

    plt.grid(True)

    plt.legend()
```

```python
def create_bar_chart_2024():

    data_2024 = df[df[col_year] == 2024]

    plt.bar(data_2024[col_month],
            data_2024[col_pages],
            color='green',
            label='Прочитано страниц')

    plt.title('Прочитанные страницы 2024')

    plt.xlabel('Месяцы')
    plt.ylabel('Количество страниц')

    plt.grid(True)

    plt.legend()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
create_line_chart_2023()

plt.subplot(1, 2, 2)
create_bar_chart_2024()

plt.tight_layout()

plt.show()
```
